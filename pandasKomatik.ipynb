# Grouping the data

```python
cereal.groupby('mfr').mean()
```

```python
cereal.groupby('mfr')['sugars'].mean()
```

```python
cereal_mfr_sugars = cereal.pivot_table(index='mfr',values='sugars')
cereal_mfr_sugars
```

```python
cereal_mfr_sugars = cereal.groupby('mfr')['sugars'].sum()

cereal_mfr_sugars.plot(kind='bar')
plt.show()
```

```python
rand_start_time = time.time()
cereal_mfr_sugars = cereal.groupby('mfr')['sugars'].mean()
print(f'ni kalo pake groupby waktunya sgini: {time.time()-rand_start_time} detik')
```
